```python
import matplotlib.pyplot as plt

def plot_with_bar(values = None, title = None, xticks= None):
    
    if (values == None):
        raise ValueError("Values no pude ser graficado")
    if (type(values) == list) or (type(values)==tuple):
        pass

    else:
        raise TypeError("values debe ser una lista table o tupla")
    
    ind = [i for i in range(len(values))]
    plt.bar(ind, values)
    plt.title(title)
    plt.xticks(ind, xticks)
    plt.show()
        
```

```python
male = dataset_numeric.query("Sex_male == 1")
female = dataset_numeric.query("Sex_female == 1")
plot_with_bar(values = [male.shape[0], female.shape[0]], title = "pasajeros por sexo", xticks = ["Hombres", "mujeres"])
```

```python
female_survived = dataset_numeric.query("Survived == 1 and Sex_female == 1")
male_survived = dataset_numeric.query("Survived == 1 and Sex_male == 1")
```

```python
plot_with_bar(values=[male_survived.shape[0], female_survived.shape[0]], title ="survived", xticks =["Hombres", "Mujeres"])
```
